```python
df["director"] = df["director"].fillna("inconnu")
df["cast"] = df ["cast"].fillna("inconnu")
df["country"] = df ["country"].fillna("inconnu")
df["rating"] = df ["rating"].fillna("inconnu")
df
```

```python
#supprimer les lignes pour date_added

df =df.dropna(subset = ["date_added"])
df
```

```python
#Transformer les colonnes multi-valeurs
df ["country"] = df["country"].str.split(",")
df ["listed_in"]= df ["listed_in"].str.split(",")
```

```python
#Créer des colonnes dérivées 

df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
#Nombre de productions par pays.

import matplotlib.pyplot as plt

top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar', color='pink')

plt.title("Top 10 des pays producteurs")
plt.xlabel("Pays")
plt.ylabel("Nombre de productions")
plt.xticks(rotation=90)
plt.show()
```

```python
#Nombre de productions par pays et par classification d’âge

import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Transformer la colonne country (car elle est sous forme de liste)
df_country = df.copy()
df_country = df_country.explode('country')

# 2️⃣ Sélectionner les 10 pays les plus producteurs
top_countries = df_country['country'].value_counts().head(10).index
filtered = df_country[df_country['country'].isin(top_countries)]

# 3️⃣ Créer le graphique
plt.figure(figsize=(10,6))
sns.countplot(data=filtered, x='country', hue='type')

plt.xticks(rotation=45)
plt.title("Nombre de productions par pays et classification d’âge")
plt.xlabel("Pays")
plt.ylabel("Nombre de productions")
plt.tight_layout()
plt.show()
```
